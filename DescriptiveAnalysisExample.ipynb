```python
import statistics
import matplotlib.pyplot as plt

member_names=["Aman","Riya","Karan","shannu","Harsha"]
ages=[22,25,24,23,26]
workout_hours=[2,1.5,2.5,1,3]
calories_burned=[500,350,600,300,750]
print("Fitness Center Dataset")
for i in range(len(member_names)):
  print(member_names[i],ages[i],workout_hours[i],calories_burned[i])


#mean workour hours
average_workout=sum(workout_hours)/len(workout_hours)
print("\nAverage workout hours:",average_workout)

#maximum Calories
print("Maximum calories: ",max(calories_burned))

#minimum calories
print("Minimum calories: ",min(calories_burned))

#median calories
print("Median calories: ",statistics.median(calories_burned))

#standard Devation
print("Standard Devation: ",statistics.stdev(calories_burned))

#most active member
index=calories_burned.index(max(calories_burned))
print("Most active member: ",member_names[index])

#bar plot
plt.bar(member_names,calories_burned)
plt.xlabel("Member")
plt.ylabel("calories Burned")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/gym_members_exercise_tracking_synthetic_data.csv'
df = pd.read_csv(file_path)

print("Fitness Center Dataset Preview:")
display(df.head())
```

```python
#average calories
average_calories = df['Calories_Burned'].mean()
print("Average Calories Burned:", average_calories)

#maximum Calories
maximum_calories = df['Calories_Burned'].max()
print("Maximum Calories Burned:", maximum_calories)

#minimum Calories
minimum_calories = df['Calories_Burned'].min()
print("Minimum Calories Burned:", minimum_calories)

#median calories
median_calories = df['Calories_Burned'].median()
print("Median Calories Burned:", median_calories)

#standard Devation
standard_deviation = df['Calories_Burned'].std()
print("Standard Deviation of Calories Burned:", standard_deviation)

```

```python
#barplotting SESSION TIME and calories
plt.scatter(df['Session_Duration (hours)'], df['Calories_Burned'])
plt.xlabel('Session_Duration (hours)')
plt.ylabel('Calories Burned')
plt.title('Session_Duration (hours) vs Calories Burned')
plt.show()
```
